# **Problem Statement**

---

The dataset analyzes air quality across cities, providing pollutant levels and AQI classifications over time. Key challenges include handling missing data and predicting AQI or AQI_Bucket categories based on pollutant metrics.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder
```

```python
df1 = pd.read_csv('city_day.csv')
df1
```

## **Dataset description**

---

1. **City**: The city where air quality data was recorded.
2. **Date**: The date of measurement.
3. Pollutants:
   
   * **PM2.5 and PM10**: Particulate
     matter with diameters less than 2.5 micrometers and 10 micrometers, respectively, which are key indicators of air quality.


   * **NO, NO2, NOx**: Nitric oxide,
     nitrogen dioxide, and their combined concentration, contributing to smog and acid rain.


   * **NH3**: Ammonia, an important
     indicator of agricultural emissions and industrial pollution.


   * **CO**: Carbon monoxide, a harmful
     pollutant from incomplete combustion.


   * **SO2**: Sulfur dioxide, which can
     form acid rain and impact respiratory health.


   * **O3**: Ozone, a pollutant at  
     ground level that contributes to smog.


  * **Benzene, Toluene, Xylene**:
     Volatile organic compounds (VOCs) that are hazardous air pollutants.


4. **AQI**: The Air Quality Index, a composite score summarizing overall air quality based on pollutant levels.

5. **AQI_Bucket**: A qualitative categorization of air quality (e.g., Good, Satisfactory, Poor).

# **Cleaning Data**

```python
#Feature Engineering

# Convert the 'date' column to datetime
df1['Date'] = pd.to_datetime(df1['Date'])

# Extract the year
df1['Year'] = df1['Date'].dt.year

# Extract the month
df1['Month'] = df1['Date'].dt.month

# Extract the day of the week (0 = Monday, 6 = Sunday)
df1['DayOfWeek'] = df1['Date'].dt.dayofweek

# droping previous column
df1.drop("Date", axis=1, inplace=True)
```

```python
df1.drop(columns = ['AQI_Bucket'], inplace = True)
```

```python
df1.drop(columns = ['Xylene'], inplace = True)
```

```python
df1['PM2.5'].fillna(df1['PM2.5'].median(), inplace = True)
df1['PM10'].fillna(df1['PM10'].median(), inplace = True)
df1['NO2'].fillna(df1['NO2'].median(), inplace = True)
df1['NOx'].fillna(df1['NOx'].median(), inplace = True)
df1['CO'].fillna(df1['CO'].median(), inplace = True)
df1['NO'].fillna(df1['NO'].median(), inplace = True)
df1['NH3'].fillna(df1['NH3'].median(), inplace = True)
df1['O3'].fillna(df1['O3'].median(), inplace = True)
df1['Benzene'].fillna(df1['Benzene'].median(), inplace = True)
df1['Toluene'].fillna(df1['Toluene'].median(), inplace = True)
df1['AQI'].fillna(df1['AQI'].median(), inplace = True)
df1['SO2'].fillna(df1['SO2'].median(), inplace = True)
```

## **Outlier Checking and Removing**

```python
# Function to remove outliers using the IQR method except for one column
def remove_outliers_iqr(df, exclude_column):
    for col in df.columns:
        # Skip the specified column
        if col == exclude_column:
            continue
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # Filter the DataFrame to remove rows with outliers in each column
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

# Specify the column to exclude
df_cleaned = remove_outliers_iqr(df1, exclude_column='DayOfWeek')
df1=df_cleaned
```

```python
x = df1[['NOx', 'PM10', 'SO2', 'NO2', 'PM2.5', 'CO']]
y = df1['AQI']

x_train, x_test, y_train, y_test = train_test_split(x, y,test_size = 0.2, random_state = 42)

model1 = LinearRegression()

model1.fit(x_train, y_train)

y_pred = model1.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)


print('Mean Squared error: ', mse)
print('r Squeared error: ', r2)
```
